```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
```

```python
sam_img = []
sam_lab = []

for i in range(10):
    idx = np.where(y_train == i)[0][0]
    sam_img.append(x_train[idx])
    sam_lab.append(class_names[i])
```

```python
plt.figure(figsize=(10, 4))

for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(sam_img[i], cmap="gray")
    plt.title(sam_lab[i])
    plt.axis("off")

plt.suptitle("One Sample Image from Each Fashion-MNIST Class")
plt.tight_layout()
plt.show()
```
